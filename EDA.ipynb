```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Dataset_Final.csv')
```

### EDA 


```python
# 1. Revisaremos los datos descriptivos del dataset, tales como la desviacion estandar, min, max
df.describe()
```

```python
# Al revisar esta informacion podemos destacar datos relevantes tales como:
# El mayor presupuesto de una pelicula tiene un valor de $3.800.000 millones de dolares.
# La mayor recaudacion tiene un valor de $2.787.000 millones de dolares
# El idioma que predomina es el ingles y el pais con la mayor produccion de peliculas es estados unidos 
```

```python
# Evaluaremos la relacion que existe entre los campos 'Revenue' y 'budget'
df = df[df['budget'] != 0]
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='budget', y='revenue')
plt.title('Relación entre Budget y Revenue')
plt.xlabel('Budget')
plt.ylabel('Revenue')
plt.show()
```

```python
# Revisaremos los outlier
# Destacar los outliers
# Veamos el boxplot de la variable 'popularity'
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, y='popularity')
plt.title('Boxplot de Popularity')
plt.ylabel('Popularity')
plt.show()
```

```python
# Generar una nube de palabras de la columna 'title'
from wordcloud import WordCloud
title_words = ' '.join(df['title'])
title_words = title_words.lower()
stopwords = ["to", "of", "the", "and", "&", ":", "in", "for", "on", "a", "by", "with", "an", "into", "from"]
wordcloud = WordCloud(stopwords=stopwords, background_color='white').generate(title_words)

plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Nube de Palabras - Títulos de Películas')
plt.show()
```

```python
#La nube de palabras nos muestra las palabras que mas resaltan en el dataset, en donde 'man', es 
# la de mayor frecuencia de aparicion, es importante debido que se puede establecer una similitud
# a la hora de recomendar una pelicula
```

```python
# Obtener los valores únicos y sus conteos en la columna 'status'

# Crear el gráfico de barras
plt.figure(figsize=(8, 5))
plt.hist(df['runtime'], bins=20, edgecolor='black')

# Etiquetas del eje x e y
plt.xlabel('Tiempo de ejecución')
plt.ylabel('Frecuencia')
plt.title('Distribución de tiempos de ejecución de películas')

# Etiquetas del eje x e y
plt.show()
```

```python
# Generar el histograma con escala logarítmica y 100 bins
plt.hist(df['vote_count'], log=True, bins=100)

# Etiqueta del eje x
plt.xlabel('Número de Votos (escala logarítmica)')
plt.ylabel('Frecuencia')
plt.title('Histograma de Conteo de Votos')

# Mostrar el histograma
plt.show()
```

Generamos un Histograma con escala logaritmica para una mejor lectura sobre distribuccion del conteo de votos sobre las peliculas
